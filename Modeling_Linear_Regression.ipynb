### Modificando o dataset:

```python
sales_mod = sales.copy()
```

Incluindo uma coluna com os anos desde a última modificação na casa.

```python
sales_mod['yr_modified'] = np.where(
                                sales_mod['yr_renovated']==0,
                                sales['yr_built'],
                                sales['yr_renovated']
                                )
```

```python
sales_mod['yr_modified'] = 2015-sales_mod['yr_modified']
```

Criando uma coluna indicando se a casa foi reformada ou não.

```python
sales_mod['renovated'] = np.where(sales_mod['yr_renovated']!=0, 1, 0)
```

```python
sales_mod.drop(
            ['yr_renovated'],
            axis=1,
            inplace=True
            )
```

Alterando a variável do porão:

```python
sales_mod['basement'] = np.where(sales_mod['sqft_basement']!=0, 1, 0)
```

```python
sales_mod.drop(
            ['sqft_basement'],
            axis=1,
            inplace=True
            )
```

Modificando as variáveis de coordenadas para o formato Ponto.

```python
from shapely.geometry import Point
```

```python
sales_mod['coordinates'] = [
                            Point(xy)
                            for xy in zip(sales_mod.long, sales_mod.lat)
                            ]
```

```python
sales_mod.drop(
            ['lat', 'long'],
            axis=1,
            inplace=True
            )
```

One hot encoding da coluna zipcode

```python
ohe = OneHotEncoder()
data = pd.DataFrame(sales_mod['zipcode'])
data = ohe.fit_transform(data)
zipcode = pd.DataFrame(
                data.toarray(),
                columns=ohe.get_feature_names_out(),
                dtype=int
                )
```

## Organizando o dataset para treinamento

```python
X = sales_mod.copy()
X.drop(
    ['zipcode'],
    axis=1,
    inplace=True
    )
X = X.join(zipcode)
```

```python
y = np.log10(X['price'])
#y = X['price']
X.drop(
    ['price'],
    axis=1,
    inplace=True
    )
```

```python
X.drop(
    ['coordinates'],
    axis=1,
    inplace=True
    )
```

# Modelos

```python
def create_model (data, target, algorithm, scaler=None):
    if scaler != None:
        X = scaler.fit_transform(data)
        #y = scaler.fit_transform(target)
    else:
        X = data
        #y = target
        
    y = target
    X_train, X_test, y_train, y_test = train_test_split(
                                                    X, #data
                                                    y, #targets
                                                    test_size=0.33,
                                                    random_state=1
                                                    )
    model = algorithm
    model.fit(X_train, y_train)

    return (model, X_train, X_test, y_train, y_test)
```

```python
scaler = RobustScaler(with_centering=False, with_scaling=True)
#scaler = MinMaxScaler()
```

```python
#regressor = LinearRegression()
#regressor = DecisionTreeRegressor()
'''
regressor = RandomForestRegressor(
                n_estimators=200,
                random_state=10,
                max_samples=0.9,
                max_features=0.8,
                max_depth=30
                )
'''

'''
regressor = xgb.XGBRegressor(
                    objective='reg:squarederror',
                    n_estimators=10000,
                    max_depth=3,
                    colsample_bytree=0.5,
                    learning_rate=0.001,
                    random_state=0,
                    subsample=0.5,
                    n_jobs=-1,
                    eval_metric='mae',
                    verbosity=0
                    )
'''

#'''
regressor = CatBoostRegressor(loss_function='RMSE',
                              iterations=10000,
                              depth=3,
                              colsample_bylevel=.4,
                              eval_metric='MAE',
                              subsample=.5,
                              random_state=0,
                              verbose=0,
                              eta=.1
                              )
#'''
```

```python
model, X_train, X_test, y_train, y_test = create_model(X, y, regressor, scaler)
#model, X_train, X_test, y_train, y_test = create_model(X, y, regressor)
```

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f'MAE: {mae:.3f}')
print(f'RMSE: {rmse:.3f}')
print(f'Score (r2 metric from the model): {model.score(X_test, y_test)}')
print(f'R2 (Test): {r2_score(y_test, y_pred)}')
print(f'R2 (Training): {r2_score(y_true=y_train, y_pred=model.predict(X_train))}')
```

```python
cv = StratifiedKFold(n_splits=10)
n_scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
print(f'R squared: {(mean(n_scores), std(n_scores))}')
```
